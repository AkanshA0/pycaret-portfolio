```python
import gradio as gr
import pandas as pd
from pycaret.classification import load_model, predict_model

model = load_model("multi_class_model")
FEATURES = list(model.feature_names_in_)  # requires sklearn>=1.0

# Build inputs as plain text/number boxes (you can customize types)
inputs = [gr.Textbox(label=col) for col in FEATURES]

def predict_from_form(*vals):
    df = pd.DataFrame([vals], columns=FEATURES)
    out = predict_model(model, data=df)
    pred = str(out.loc[0, "prediction_label"])
    prob_cols = [c for c in out.columns if "Probability" in c]
    probs = {c.replace("Probability_", ""): float(out.loc[0, c]) for c in prob_cols}
    return {"prediction": pred, "probabilities": probs}

demo = gr.Interface(
    fn=predict_from_form,
    inputs=inputs,
    outputs=gr.JSON(label="Result"),
    title="PyCaret Multiclass Classification Demo",
    allow_flagging="never",
)

if __name__ == "__main__":
    demo.launch()
```
